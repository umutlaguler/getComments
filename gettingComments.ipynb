```python
#library that we will use for the get comments from Hepsiburada.com
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
#import for write comments into a csv file
import pandas as pd

options=webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
#chromedriver very important part of the project to open a browser clone with selenium lib.
chrome_path=r'/Users/umutguler/bin/chromedriver'
driver=webdriver.Chrome(chrome_path)

#to store all comments into an array
comment_list=[]
#to get all comments from a single page 
def getComments():
    i = 1
    divNumber = 2
    for divNumber in range(11):
        #get comments by xpath 
        comments=driver.find_elements_by_xpath("//*[@id='hermes-voltran-comments']/div[6]/div[3]/div[1]/div[{}]/div[2]/div[2]/span".format(divNumber))
        for c in comments:
            #append all coments into comment_list which has already been declared
            comment_list.append(c.text+"\n")
            #to print all comments (optional)
            print(c.text)
#get the first pages's comments
driver.get("https://www.hepsiburada.com/oppo-a74-128-gb-oppo-turkiye-garantili-p-HBCV0000097C74-yorumlari")
#call getComments function to repeat it
getComments()
#counter for change the page URL to get other pages's comments
j = 2
#50 pages that we will get comments
while j<=50:
    #changing URL 
    driver.get("https://www.hepsiburada.com/oppo-a74-128-gb-oppo-turkiye-garantili-p-HBCV0000097C74-yorumlari?sayfa={}".format(j))
    #to make it reliable we added waiting time 
    time.sleep(20)
    j += 1
    #calling getComments function to repeat the same process
    getComments()
#to make it reliable we added waiting time     
time.sleep(10) 

#to write comments into a csv file 
df=pd.DataFrame({"comments":comment_list})
df.to_csv('comments.csv',index=False, encoding='utf-8-sig')

#to write into TXT file which is optional  
# text_file=open("tech_comments.txt","w", encoding="UTF-8")
# for c in comments:
#     text_file.write(c.text+"\n")
#     print(c.text)

# text_file.close()
```
